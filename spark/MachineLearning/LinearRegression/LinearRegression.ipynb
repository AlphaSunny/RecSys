```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("lr_regression").getOrCreate()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
training = spark.read.format("libsvm").load("sample_linear_regression_data.txt")
```

```python
training.show()
```

```python
lr = LinearRegression(featuresCol='features',  labelCol='label', predictionCol='predictions')
```

```python
lrModel = lr.fit(training)
```

```python
print('coefficients: {}'.format(lrModel.coefficients))
print('\n')
print("intercepts: {}".format(lrModel.intercept))
```

```python
trainingSummery = lrModel.summary
```

```python
trainingSummery.residuals.show()
print("r2: {}".format(trainingSummery.r2))
print("rmse:{}".format(trainingSummery.rootMeanSquaredError))
```

# train test split

```python
all_data = spark.read.format("libsvm").load("sample_linear_regression_data.txt")
```

```python
train_data, test_data = all_data.randomSplit([0.7, 0.3])
```

```python
train_data.show()
```

```python
test_data.show()
```

```python
unlabeled_data = test_data.select('features')
```

```python
unlabeled_data.show()
```

```python
correct_model = lr.fit(train_data)
```

```python
test_results = correct_model.evaluate(test_data)
```

```python
test_results.residuals.show()
print("Rmse: {}".format(test_results.rootMeanSquaredError))
```

```python
predictions = correct_model.transform(train_data)
```
